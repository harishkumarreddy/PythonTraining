# Randomes

```
a1 = np.zeros([3,3])
print(a1)

a2 = np.ones([3,3])
print(a2)

a3 = np.full([3,3], 10)
print(a3)

a4 = np.identity(3)
print(a4)

a5 = np.random.rand(3, 3)
print(a5)

a6 = np.random.randint(1, 10, [3,3])
print(a6)
```

```
a6 + a4
a6.size
```

```
a6_d = a6
print(a6)
print(a6_d)

a6_d[1,1] = 10

print(a6_d)
print(a6)

a6_c = a6.copy()
a6_c[1,1] = 20

print(a6_c)
print(a6)
```

```
a6_d = a6.view()
print(a6)
print(a6_d)
print("=" * 20)
a6_d[1,1] = 15

print(a6_d)
print(a6)
print("=" * 20)
a6[1,1] = 20

print(a6_d)
print(a6)
```

```
arr1 = np.array([[1, 2, 3], [4, 5, 6]])

arr2 = np.array([[4, 5, 6], [1, 2, 3]])

arr = np.concatenate((arr1, arr2))
print(arr)

arr = np.concatenate((arr1, arr2), axis=1)
print(arr)
```

```
arr1 = np.array([3, 2,  5,4,1, 6])
print(arr1)

arr2 = np.array_split(arr1,3)
print(arr2)

print(np.where(arr1 == 4))

print(np.sort(arr1))

print(arr1[arr1%2 == 1])

print(arr1[arr1 >=5])
```
